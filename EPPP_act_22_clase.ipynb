####clase gerente

```python
# Definición de la clase Gerente
class Gerente:
    # El constructor permite asignar nombre y departamento al momento de crear el objeto
    def __init__(self, nombre, departamento):
        self.nombre = nombre
        self.departamento = departamento

    # Función para que el gerente dé órdenes (sustituye a ladrar)
    def dar_ordenes(self):
        print(f"El gerente {self.nombre} está dando órdenes: ¡Necesito el reporte para ayer! 📈")

    # Función para que el gerente asista a juntas (sustituye a correr)
    def asistir_junta(self):
        print(f"El gerente {self.nombre} del departamento de {self.departamento} está en una junta de Zoom 💻")

    # Función para que el gerente despida (sustituye a morder)
    def despedir(self):
        print(f"El gerente {self.nombre} está despidiendo a Recursos Humanos por error 🤦‍♂️")

# --- Creación de los objetos con datos específicos ---

# Objeto 1: Gerente de Ventas
gerente_ventas = Gerente("Rocco", "Ventas")

# Objeto 2: Gerente de Finanzas
gerente_finanzas = Gerente("Pipo", "Finanzas")

# Objeto 3: Gerente de Proyectos
gerente_proyectos = Gerente("Solovino", "Proyectos")

# --- Ejecución de las funciones ---

# Acciones para el Gerente de Ventas
gerente_ventas.dar_ordenes()
gerente_ventas.asistir_junta()

print("-" * 30) # Separador visual en la consola

# Acciones para el Gerente de Finanzas
gerente_finanzas.dar_ordenes()
gerente_finanzas.asistir_junta()

print("-" * 30) # Separador visual en la consola

# Acciones para el Gerente de Proyectos
gerente_proyectos.dar_ordenes()
gerente_proyectos.asistir_junta()
gerente_proyectos.despedir()

print("\nPrograma realizado por Ambar Fierro NL = 19")
```
